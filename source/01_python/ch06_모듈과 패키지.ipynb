````python
# 1. import [모듈] : 모듈 안의 파일(함수, 클래스, 변수)을 메모리에 로드.

import time # time 모듈 안의 모든 파일을 메모리에 로드
print(time.ctime()) # 현재 시각과 날짜
print(time.time()) #1970.1.1부터의 초
````

````python
import time
start = time.time()
cnt = 0
while cnt<=10000000:
    cnt += 1
end = time.time()
print('반복문 1천만번 돈 시간 : ',(end-start), '초')
````

````python
del time # del [모듈] : 모듈을 메모리에서 언로드.
````

````python
# 2. from [모듈] import [파일] : 모듈 내 특정 파일(함수, 클래스, 변수) 로드
#    from [패키지] import [모듈] : 패키지 내 특정 모듈 로드
#    from [패키지].[모듈] import [파일] : 패키지 내 특정 모듈의 특정 파일 로드

from time import ctime, time
ctime()
````

````python
del ctime, time
````

````python
# 3. import [모듈] as [별칭] : 모듈을 별칭으로 로드
        # 고정 별칭 예시) import numpy as np

import time as t
t.ctime()
````

````python
del t
````

````python
# 4. from [모듈] import [파일] as [별칭] : 모듈의 특정 파일을 별칭으로 로드
#    from [패키지] import [모듈] as [별칭] : 패키지 내 특정 모듈을 별칭으로 로드
#    from [패키지].[모듈] import [파일] as [별칭] : 패키지 내 특정 모듈의 특정 파일을 별칭으로 로드

from time import ctime as ct, time as ti
ct()
````

````python
import time
print(dir(time)) # time 모듈 내 파일을 리스트로 보여줌
````

```
모듈이나 모듈 내 함수를 가져오는 방법

import A               - A 로드
import A as B          - A를 B 별칭으로 로드
from A import B        - A 모듈 내 B 파일 로드
from A import B as C   - A 모듈 내 B 파일을 C 별칭으로 로드

from A import B        - A 패키지 내 B 모듈 로드
from A.B import C      - A 패키지 내 B 모듈의 C 파일 로드
from A import B as C   - A 패키지 내 B 모듈을 C 별칭으로 로드
from A.B import C as D - A 패키지 내 B 모듈의 C 파일을 D 별칭으로 로드
```
