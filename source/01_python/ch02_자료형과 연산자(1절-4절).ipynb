#### 정규 표현식(regex)
```
    1. 참조 : https://ko.wikipedia.org/wiki/정규_표현식
    2. 간략문법
            \d : 숫자와 매치; [0-9]와 동일
            \D : 숫자가 아닌 문자와 매치
            \w : 영문자(대소문자) or 숫자와 매치; [a-zA-Z0-9]와 동일
            \W : 영문자 or 숫자가 아닌 문자
            .  : 문자
            \. : .
            \s : [ ](공백)
            ?  : 0~1회 반복
            +  : 1회 이상 반복
            *  : 0회 이상 반복
            {n,} : n회 반복 ex) 숫자 n자리 연속된 것만 분류
            {n,m}: n~m회 반복 ex) 숫자 n자리~m자리 연속된 것만 분류
    3. 정규 표현식 연습장 : https://regexr.com/
        ex1. 전화번호 패턴 : [0]\d{1,2}[\-)\s\.]\d{3,4}[\-)\s\.]\d{4}
        ex2. 이메일 패턴 : 내 답 : \w+@\w+\.\w+\.?\w+ , 답안 : \w+@\w+(\.\w+){1,2}
```

````python
data="이름 : 홍길동, 주소 : 신림동, 전화 : 02-888-8888, \
휴대폰 : 010-9999-9999, 이메일 : aabc_1234@mbc.co.kr, git__hub2@gmail.com \
특징 : 서해번쩍 동해번쩍, 혈액형 = B ㅋㅋㅎ"
````

````python
import re
tel_pattern = r"[0]\d{1,2}[\-)\s\.]\d{3,4}[\-)\s\.]\d{4}"
tels = re.findall(tel_pattern, data) #해당 패턴을 모두 찾고, 없을 경우 빈 list([])를 나타냄.
tels
````

````python
#정규표현식 패턴과 처음 매치되는 것을 찾아줘 re.search([패턴], [내용])
firstTel = re.search(tel_pattern, data)
firstTel
````

````python
print(firstTel.group(0)) # 0번째 인덱스 그룹
print(data[25:36])
````

````python
phone_pattern = r"(\d{2,3})-(\d{3,4})-(\d{4})"
phone = re.search(phone_pattern, data)
print(phone.group(0))
print(phone.group(1))
print(phone.group(2))
print(phone.group(3))
````

````python
import re
email_pattern = "\w+@\w+\.\w+\.?\w+"
emails = re.findall(email_pattern, data)
emails
````

````python
firstemail = re.search(email_pattern, data)
firstemail
````

````python
emailID_pattern = "(\w+)@(\w+\.\w+\.?\w+)"
ID = re.search(emailID_pattern, data)
print(ID.group(1))
````

````python
#특정 문자만 없애기
re.sub(r"[^가-힣0-9\-a-zA-Z@._]"," ", data)
````

## 4.2 format()

````python
# %s : 문자, %d : 10진수 정수, %f : 실수
print("%s님 나이는 %d세 입니다."%(name, age))
print(f"{name}님 나이는 {age}살 입니다.")
````

````python
print("{}님 나이는 {}세 입니다.".format(name, age))
print("{0}님 나이는 {1}세 입니다.".format(name, age)) #format() 내의 각각 인덱스의 0번째, 1번째가 들어감.
print("{1}님 나이는 {0}세 입니다.".format(age, name))
print("{0}세 {1:4}님 반갑습니다. {1}님 어서오세요".format(age, name)) # 1번째를 4칸을 확보하여 출력함.
````

- "{[인덱스]:[공백대체문자][정렬방법][자릿수][타입]}내용".format(변수명)
    - 인덱스 : format() 함수 안의 매개변수 순서
    - 공백대체문자는 저렬방식과 자릿수를 명시하여야 함
    - 정렬방법 : 정렬할 방법을 지정
        - '<' : 왼쪽 정렬
        - '>' : 오른쪽 정렬
        - '^' : 가운데 정렬
    - 정수 자릿수 : 변수의 값을 출력할 최대 자릿수를 지정
    - 실수 자릿수 : "(전체 출력될 자릿수).(변수 중 출력할 자릿수)"
    - 타입 : 출력 형식 지정
        - d(10진 정수), f(실수 ex) 4.1f:1은 소수점자릿수), b(2진수), o(8진수), x(16진수), s(문자)

````python
name = input("이름을 입력해주세요.")
age = int(input("나이를 입력해주세요."))
print("{0}세 {1}님 안녕하세요.".format(age, name))
print(f"{age}세 {name}님 안녕하세요.")
````

````python
name = "홍길동"
age = 30
print("[{0:*^10}]세 [{1:*^10}]님 안녕하세요.".format(age, name))
print(f"[{age:_>3}]세 [{name:*<3.1}]님 안녕하세요.")
````

````python
a = 2.75544
sign = '+'
width = 10
digits = 4
print('{:+10.4}'.format(a))
print(f'{a:{sign}{width}.{digits}}')
print('{:{sign}{width}.{digits}}'.format(a,
                            sign=sign,
                            width=width,
                            digits=digits))
````

## 4.3 날짜 출력

````python
from datetime import datetime

day = datetime(2026,11,19,18,0,0) #날짜형 데이터는 날짜와 시간이 모두 포함됨

print(day)
````

````python
# Y (2026) y(26) h(Nov) m(11) D(11/19/26) p(am/pm) I(12h) H(24h)
print("{:%Y년 %m월 %d일 %p %I시}".format(day))
print(day.strftime("%Y년 %m월 %d일 %p %I시"))
````

````python
from time import localtime, strftime
today = localtime() #현재의 날짜와 시간을 표기하는 함수
now = strftime("%Y년 %m월 %d일 %p %I시 %M분".format(today))
print(now)
````
