# 2. try ~ except 
## 2.1 try~except 예외처리
```
try:
    예외가 발생할 가능성이 있는 명령어
except:
    예외가 발생할 경우 실행할 명령어
```

````python
# 100을 입력받은 정수값으로 나눈 결과를 출력

x = int(input('정수를 입력하세요 : '))
# ValueError 예외 객체의 메세지 : invalid literal for int() with base 10: 'a'
#                              a를 입력했을 때의 예외 타입 (형변환 불가한 문자 입력)
print(f'입력한 정수는 {x} 입니다.')
print(f'100을 {x}로 나눈 결과 : {100/x:.2f}')
# ZeroDivisionError 예외 객체의 메시지 : division by zero
#                                     0을 입력했을 때의 예외 타입 (0으로 나누기 불가)
````

````python
# 100을 입력받은 정수값으로 나눈 결과를 출력하기

try:
    x = int(input('정수를 입력하세요 : '))
    print(f'입력한 정수는 {x} 입니다.')
    print(f'100을 {x}로 나눈 결과 : {100/x:.2f}')

except ValueError:
    print('유효한 정수가 아닙니다.')
except ZeroDivisionError: # 예외가 발생할 경우 except가 실행. 0 입력 시 '입력한 정수는 0 입니다.'는 출력됨.
    print('유효한 정수가 아닙니다.')
    
print('Done') # except 이후의 코드도 모두 실행됨.
````

````python
# 100을 나눌 유효한 정수를 입력할 때까지 정수를 입력받아 유효한 정수를 입력하면 입력한 정수와 100으로 나눈 수를 출력하기
while True:
    try:
        x = int(input('정수를 입력하세요 : '))
        print(f'입력한 정수는 {x} 입니다.')
        print(f'100을 {x}로 나눈 결과 : {100/x:.10f}')
        break

    except:
        print('유효한 정수가 아닙니다.')
````

## 2.2 예외를 지정한 처리
```
try:
    예외가 발생할 수도 있는 명령어
except 예외타입1:
    해당 예외가 발생할 경우 실행할 명령어
except 예외타입2:
    해당 예외가 발생할 경우 실행할 명령어
```

**예외 지정이 많을 경우, 상위 예외타입(상속관계의 예외타입)은 하위 예외타입의 아래로 위치해야함**

````python
# 100을 나눌 유효한 정수를 입력할 때까지 정수를 입력받아 유효한 정수를 입력하면 입력한 정수와 100으로 나눈 수를 출력하기
while True:
    try:
        x = int(input('정수를 입력하세요 : '))
        print(f'입력한 정수는 {x} 입니다.')
        print(f'100을 {x}로 나눈 결과 : {100/x:.2f}')
        break

    except ZeroDivisionError: # 예외 지정 1
        print('0으로 나눌 수 없습니다. 다시 입력하세요.')
        
    except ValueError: # 예외 지정 2
        print('정수가 아닙니다. 다시 입력하세요.')
        
    except: # 그 외의 모든 예외 발생 시
        print('예상하지 못한 오류가 발생했습니다. 다시 입력하세요.')
        
print('DONE')
````

````python
# 100을 나눌 유효한 정수를 입력할 때까지 정수를 입력받아 유효한 정수를 입력하면 입력한 정수와 100으로 나눈 수를 출력하기
while True:
    try:
        x = int(input('정수를 입력하세요 : '))
        print(f'입력한 정수는 {x} 입니다.')
        print(f'100을 {x}로 나눈 결과 : {100/x:.2f}')
        break

    except (ZeroDivisionError, ValueError): # 예외 지정 시 튜플 사용 가능
        print('나눗셈이 불가능한 정수입니다. 다시 입력하세요.')
    
    except: # 그 외의 모든 예외 발생 시
        print('예상하지 못한 오류가 발생했습니다. 다시 입력하세요.')
        
print('DONE')
````

### try ~ except ~ finally
```
try:
    예외가 발생할 수도 있는 명령어
except [예외 타입 as e]:
    예외가 발생할 경우 실행할 명령어
[else:
    try 블록을 모두 수행한 후 실행할 명령어] - 자주 사용하지 않음
finally:
    try 블록 또는 except 블록을 모두 수행한 후 실행할 명령어
```

````python
try:
    f = open('test/ch08.txt','r') # encoding='utf-8' # 파일 경로 및 오픈 방식 지정 (파일 열기)
    data = f.read() # 파일 읽기

except (FileNotFoundError, UnicodeDecodeError) as e:
    print(e)
    
else:
    print(data) # 파일을 이용한 함수
    
finally:
    f.close() # 파일 경로 삭제 (파일 닫기)
````

````python
try:
    f = open('test/ch08.txt','r')
    data = f.read()
    print(data)
    
except (FileNotFoundError, UnicodeDecodeError) as e:
    print(e)

finally: # try 블록 수행 또는 except 블록 수행 후 실행
    f.close()
````

````python
# 사용자 정의 예외 : Exception 클래스를 상속 받은 클래스
class LengthZeroError(Exception):
        'len() 결과가 0일 때 발생하는 예외'
        def __init__(self, message):
            super().__init__(message) # 부모 클래스의 생성자에서 예외 메세지를 세팅
````

````python
raise LengthZeroError("len() 결과가 0 입니다")
````

````python
def args_print(*data):
    if len(data)==0:
        raise LengthZeroError("len() 결과가 0 입니다 (사용자 정의 예외)")
    for item in data:
        print(item, end=' ')
````

````python
args_print(1, 2, 3, 4, 5)
````

````python
args_print()
````

````python
try:
    args_print()
except Exception as e:
    print(e)
````

````python
# ch04_제어문 - 중첩 루프 탈출 방법 2 : Exception(예외처리) 활용

class BreakOutLoopError(Exception):
    pass

try:
    for a in range(0, 3):
        for b in range(1, 3):
            if a==b:
                raise BreakOutLoopError() # 강제 예외 발생
            print(a, b)
except BreakOutLoopError:
    pass

print('Done')
````

# 4. 추상 클래스

- 추상 클래스 : 추상 메소드가 하나 이상 포함된 객체를 생성할 수 없는 클래스
    * 자식 클래스들이 반드시 구현(재정의)해야 하는 메서드의 구조를 강제함
    * 프로그램의 공통 규격(인터페이스)을 맞추기 위해 사용하는 미완성 설계도
- 파이썬의 추상 클래스 구현 : `raise` or `@abstractmethod`
- 추상 메소드 : 내용(코드 body)이 비어 있는 메소드
    * 메소드의 이름과 역할을 선언만 해두고, 실제 구체적인 기능은 구현하지 않음
    
**규격화와 강제성, 코드의 재사용성, 중복 방지 등의 역할**

````python
#1 raise NotImplementedError 를 이용한 추상 클래스 구현

class Shape: #부모클래스
    def __init__(self):
        raise NotImplementedError('추상 클래스')
    def calc_area(self):
        raise NotImplementedError('추상 메소드') # 자식 클래스가 오버라이딩 없이 호출 시 에러 강제 발생
````

````python
class Circle(Shape): #자식클래스
    
    def __init__(self, x, y, r=3):
        self.x = x
        self.y = y
        self.r = r
        
    def calc_area(self):
        import numpy as np
        print(f'{self.r*self.r*np.pi}')
        
    def __str__(self):
        import numpy as np
        return f'좌표 : {self.x}, {self.y}, 넓이 :{self.r*self.r*np.pi}'
````

````python
myCircle = Circle(3,3,5)
print(myCircle)
````

````python
#2 @abstractmethod를 이용한 추상 클래스 구현
from abc import ABC, abstractclassmethod

class Shape(ABC):
    @abstractclassmethod
    def __init_(self):
        pass
    @abstractclassmethod
    def calc_area(self):
        pass
````

````python
class Circle(Shape):
    def __init_(self):
        pass
````

````python
myCircle = Circle() #부모 클래스의 추상 클래스를 자식 클래스에서 구현하지 않을 시 에러 발생
````

````python
def isit_int(a):
    return int(a) if a%1==0 else a

while True:
    try:
        x = isit_int(float(input('첫 번째 숫자를 입력하세요 : ')))
        y = isit_int(float(input('두 번째 숫자를 입력하세요 : ')))
        print(f'입력한 수는 {x}와 {y} 입니다.')
        value = isit_int(x/y)
        print(f'{x}을 {y}로 나누면 {value} 입니다.')
        break
        
    except ZeroDivisionError:
        print('0으로 나눌 수 없습니다. 다시 시도하세요.')        
    
    except ValueError:
        print('유효한 숫자가 아닙니다. 다시 시도하세요.')
````
