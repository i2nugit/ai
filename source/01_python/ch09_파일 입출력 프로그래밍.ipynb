## 2.1 형식이 있는 텍스트 데이터 불러오기
- ch09_member.txt → 객체 리스트 변환

````python
with open('data/ch09_member.txt', 'rt', encoding='utf-8') as f:
    lines = f.readlines()
    
for line in lines:
    data = line.strip().split(', ') # strip() : space, enter, tab 등을 제거함
    name = data[0]
    age = int(data[1])
    email = data[2]
    address = data[3]
    print("이름 : {}, 나이 : {}, 이메일 : {}, 주소 : {}".format(name, age, email, address))
````

````python
class Member:
    def __init__(self, name, age, email, address):
        self.name = name
        self.age = age
        self.email = email
        self. address = address
        
    def __eq__(self, other):
        if isinstance(other, Member):
            return self.__str__() == other.__str__() # a == b 사용 시, a.__eq__(b)가 호출됨.
        return False
    
    def __str__(self):
        return f"{self.name}, {self.age}, {self.email}, {self.address}"
    
#    def as_dict(self): # 객체를 딕셔너리로 반환
#        return {'name':self.name, 'age':self.age, 'email':self.email, 'address':self.address}
    
    def as_list(self): # 객체를 리스트로 반환
        return [self.name, self.age, self.email, self.address]
````

````python
member = Member("홍길동","33","hongil@h.co.kr","신림동")
print(member)
````

````python
member = Member("홍길동","33","hongil@h.co.kr","신림동")
member2 = Member("홍길동","33","hongil@h.co.kr","신림동")
print(id(member), id(member2))
print(member==member2)
print(member.__eq__(member2))
print(member.__dict__)
print(member.as_list())
````

````python
# 파일의 내용을 member list나 dict list로 만들기

member_list=[] # 파일의 내용이 member 객체로 추가될 빈 리스트
member_dict_list = []

with open('data/ch09_member.txt', 'rt', encoding='utf-8') as f:
    lines = f.readlines()
    
for line in lines:
    data = line.strip().split(',') # strip() : space, enter, tab 등을 제거함
    name = data[0].strip()
    age = int(data[1].strip())
    email = data[2].strip()
    address = data[3].strip()
    member = Member(name, age, email, address)
    member_list.append(member)
    member_dict_list.append(member.__dict__)
````

### 객체 리스트를 json 파일로 저장 시

````python
 
class Member:
    def __init__(self, name, age, email, address):
        self.name = name
        self.age = age
        self.email = email
        self.address = address
    def as_dict(self):
        '객체를 딕셔너리로 바꾸어 반환, json.dump 사용 시 필요'
        return self.__dict__
    def __str__(self):
        return f'이름 :{self.name}, 나이 :{self.age}, 이메일 :{self.email}, 주소 :{self.address}'
    def __eq__(self, other):
        if isinstance(other, Member): # other가 Member class일 경우, return
            return self.__str__() == other.__str__()
        else:
            return False
````

````python
member1 = Member('홍길동',20,'h@h.com','관악구')
member2 = Member('홍길동',20,'h@h.com','관악구')
print(member1 == member2) # a == b 실행 시 a.__eq__(b) 호출.
print(member1.__eq__("홍길동"))
````

````python
member_list = [Member('홍길동',20,'h1@h.com','신대방'),
              Member('김길동',30,'h2@h.com','봉천'),
              Member('신길숙',25,'h3@h.com','신림')]
for member in member_list:
    print(member)
````

````python
# default : JSON으로 변환할 수 없는 객체를 만났을 때, 이를 어떻게 처리할지 알려줌. 함수 지정 가능.
with open('data/ch09_member2.json', 'w', encoding='utf-8') as jsonf:
    json.dump(member_list, jsonf, ensure_ascii=False, indent=' ', 
              default=Member.as_dict) # 객체를 딕셔너리 형태로 반환하는 인스턴스 함수 지정
````

```
홍길동, 010-9999-9999, k@hong.com, 30, 5, vvip
유길동, 010-5432-2222, y@hong.com, 20, 2, 가칠해
```

### Customer class 작성
    * Customer("홍길동", "010-9999-9999", "k@hong.com", 30, 5, "vvip")
    * __init__, __str__, as_dict, to_txt_style(객체를 txt 파일로 출력)

````python
class Customer:
    def __init__(self, name, phone, email, age, grade, etc):
        self.name  = name
        self.phone = phone
        self.email = email
        self.age   = age
        self.grade = grade
        self.etc   = etc
    def as_dict(self):
        return self.__dict__
    
    def __str__(self): # ***  홍길동    010-9999-9999  k@k.com   30   vvip
        return "{:>6}\t{:>5}\t{:15}\t{:^20}\t\t{:>3}\t{}".format(self.grade*'*',
                                                           self.name, 
                                                           self.phone,
                                                           self.email,
                                                           self.age,
                                                           self.etc)
    
    def to_text_style(self): # 홍길동, 010-9999-9999, k@hong.com, 30, 5, vvip
        return "{}, {}, {}, {}, {}, {}".format(self.name, 
                                              self.phone,
                                              self.email,
                                              self.age,
                                              self.grade,
                                              self.etc)
````

````python
def to_customer(row):
    data = row.strip().split(', ')
    name = data[0]
    phone = data[1]
    email = data[2]
    age   = int(data[3])
    grade = int(data[4])
    etc   = data[5]
    return Customer(name, phone, email, age, grade, etc)
````

### 첫 실행 시
```
data/ch09_custeomers.txt의 내용을 customer_list에 입력받고 customer_list를 return
data/ch09_custeomers.txt 빈 파일 생성하고 customer_list=[], customer_list를 return
```

````python
def load_customers():
    customer_list = []
    try:
        with open('data/ch09_customers.txt', 'r', encoding='utf-8') as f:
            lines = f.readlines()
            for line in lines:
                customer = to_customer(line)
                customer_list.append(customer)
    except FileNotFoundError:
        print("데이터 파일 ch09_customers.txt는 존재하지 않는 파일입니다. 파일을 생성합니다.")
        with open('data/ch09_customers.txt', 'w', encoding='utf-8') as f:
            f.write('')
    
    return customer_list
````

### 정보 입력

````python
def fn1_insert_customer_info():
    import re
    
    name = input('이름 : ')
    name_pattern = r'[가-힣]{2,}'
    while not re.search(name_pattern, name):
        print('알 수 없는 이름입니다.')
        name = input('이름 : ')

    phone = input('전화번호 : ')
    tel_pattern = r"[0]\d{1,2}[\-)\s\.]\d{3,4}[\-)\s\.]\d{4}"
    while not re.search(tel_pattern, phone):
        print('올바른 전화번호 형식이 아닙니다. -를 포함하여 입력하세요.')
        phone = input('전화번호 : ')   

    email = input('이메일 : ')
    email_pattern = r"(\w+)@(\w+\.\w+\.?\w+)"
    while not re.search(email_pattern, email):
        print('올바른 이메일 형식이 아닙니다.')
        email = input('이메일 : ') 

    while True:
        try:
            age = int(input('나이 : '))
            if (age<=0)|(130<age):
                raise Exception('잘못된 입력입니다.')
            break
        except:
            print("올바른 나이를 입력하세요")
            
    try:
        grade = int(input('고객등급 : '))
        if grade<1:
            grade = 1
            print("1 등급으로 입력됩니다.",)
        if grade>5:
            grade = 5
            print("5 등급으로 입력됩니다.")
    except:
        grade = 1
        print("올바른 등급을 입력하지 않을 시 1 등급으로 입력됩니다.")
        
    etc = input('etc : ')
    print()
    return Customer(name, phone, email, age, grade, etc)
````

### 정보 조회

````python
def fn2_print_customers(customer_list):
    print("="*100)
    print("{:^100}".format('고객정보'))
    print("-"*100)
    print("{:>5}{:>4}\t{:^25}\t{:15}\t{:>3}\t{}".format('고객등급',
                                                          '이름', 
                                                          '전화번호',
                                                          '이메일',
                                                          '나이',
                                                          '기타'))
    print("-"*100)
    for customer in customer_list:
        print(customer)
````

### 실행 종료

````python
def fn3_save_customer_txt(customer_list):
    customer_text_list = []
    for customer in customer_list:
        text = customer.to_text_style()+'\n'
        customer_text_list.append(text)
    with open('data/ch09_customers.txt', 'w', encoding='utf-8') as f:
        f.writelines(customer_text_list)
    print('백업 완료, 프로그램을 종료합니다.')
````

````python
def main():
    global customer_list
    customer_list = load_customers()
    while True:
        try:
            while True:
                print("1:입력", "2:전체 출력", "3:데이터 백업 후 종료 ", sep=' | ', end='')
                fn = int(input('* 메뉴 선택 : '))
                if fn == 1:
                    customer = fn1_insert_customer_info()
                    customer_list.append(customer)
                if fn == 2:
                    fn2_print_customers(customer_list)
                if fn == 3:
                    fn3_save_customer_txt(customer_list)
                    break
            break
        except:
            print('다시 입력하세요.')
````

````python
if __name__ == '__main__':
    main()
````
