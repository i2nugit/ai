# 1. txt 파일 입출력

* 작업 종류
    - w : 기존의 파일이 있는 경우 덮어쓰기. 없을 경우 파일 생성.
    - a : 기존 내용을 보존, 파일의 맨 끝에 내용을 이어 붙임. 없을 경우 파일 생성.
    - r : 기본값, 파일을 읽기만 할 때 사용. 없을 경우 FileNotFoundError 에러 발생.
    - x : 파일을 새로 만들어 쓰기 모드로 생성. 기존의 파일이 있는 경우 FileExistsError 에러 발생.

* 파일 형식 지정 - t : 텍스트 모드, 기본값 (.txt, .csv)

# 2. pickle 객체 입출력

## 2.1 형식이 있는 텍스트 데이터 불러오기
- ch09_member.txt → 객체 리스트 변환

```python
with open('data/ch09_member.txt', 'rt', encoding='utf-8') as f:
    lines = f.readlines()
    
for line in lines:
    data = line.strip().split(', ') # strip() : space, enter, tab 등을 제거함
    name = data[0]
    age = int(data[1])
    email = data[2]
    address = data[3]
    print("이름 : {}, 나이 : {}, 이메일 : {}, 주소 : {}".format(name, age, email, address))
```

```python
class Member:
    def __init__(self, name, age, email, address):
        self.name = name
        self.age = age
        self.email = email
        self. address = address
        
    def __eq__(self, other):
        if isinstance(other, Member):
            return self.__str__() == other.__str__() # a == b 사용 시, a.__eq__(b)가 호출됨.
        return False
    
    def __str__(self):
        return f"{self.name}, {self.age}, {self.email}, {self.address}"
    
#    def as_dict(self): # 객체를 딕셔너리로 반환
#        return {'name':self.name, 'age':self.age, 'email':self.email, 'address':self.address}
    
    def as_list(self): # 객체를 리스트로 반환
        return [self.name, self.age, self.email, self.address]
```

```python
member = Member("홍길동","33","hongil@h.co.kr","신림동")
print(member)
```

```python
member = Member("홍길동","33","hongil@h.co.kr","신림동")
member2 = Member("홍길동","33","hongil@h.co.kr","신림동")
print(id(member), id(member2))
print(member==member2)
print(member.__eq__(member2))
print(member.__dict__)
print(member.as_list())
```

```python
# 파일의 내용을 member list나 dict list로 만들기

member_list=[] # 파일의 내용이 member 객체로 추가될 빈 리스트
member_dict_list = []

with open('data/ch09_member.txt', 'rt', encoding='utf-8') as f:
    lines = f.readlines()
    
for line in lines:
    data = line.strip().split(',') # strip() : space, enter, tab 등을 제거함
    name = data[0].strip()
    age = int(data[1].strip())
    email = data[2].strip()
    address = data[3].strip()
    member = Member(name, age, email, address)
    member_list.append(member)
    member_dict_list.append(member.__dict__)
```

```python
[member.__str__() for member in member_dict_list]
```

## 2.2 피클링
- member_list → 피클 파일에 쓰기
- 피클 파일을 읽기 → loaded_member_list
- 파일 형식 지정 - b : 바이너리 모드 (.pkl)


- 문법
   - 피클링
   `pickle.dump(저장하고 싶은 객체, 피클링할 파일 객체)`
   - 언피클링
   `pickle.load(객체로 불러올 피클링 파일)`

```python
# 피클링(확장자: pkl, pickle, data)을 이용하여 객체를 파일에 저장하기

import pickle
with open('data/ch09_member.pkl', 'wb') as f:
    pickle.dump(member_list, f)
```

```python
# 피클링을 이용하여 파일에서 객체 데이터 읽기

with open('data/ch09_member.pkl', 'rb') as f:
    loaded_member_list = pickle.load(f)
```

```python
# member_list와 loaded_member_list 비교
for member, loaded_member in zip(member_list, loaded_member_list):
    print(member,'/',loaded_member,sep='')
```

```python
# member_list와 loaded_member_list 비교 2 (__eq__ 처리 이후)
for member, loaded_member in zip(member_list, loaded_member_list):
    print(member == loaded_member)
```
