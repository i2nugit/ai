# 2. try ~ except 예외처리
## 2.1 try~except
```
try:
    예외가 발생할 가능성이 있는 명령어
except:
    예외가 발생할 경우 실행할 명령어
```

````python
# 100을 입력받은 정수값으로 나눈 결과를 출력

x = int(input('정수를 입력하세요 : '))
# ValueError 예외 객체의 메세지 : invalid literal for int() with base 10: 'a'
#                              a를 입력했을 때의 예외 타입 (형변환 불가한 문자 입력)
print(f'입력한 정수는 {x} 입니다.')
print(f'100을 {x}로 나눈 결과 : {100/x:.2f}')
# ZeroDivisionError 예외 객체의 메시지 : division by zero
#                                     0을 입력했을 때의 예외 타입 (0으로 나누기 불가)
````

````python
# 100을 입력받은 정수값으로 나눈 결과를 출력하기

try:
    x = int(input('정수를 입력하세요 : '))
    print(f'입력한 정수는 {x} 입니다.')
    print(f'100을 {x}로 나눈 결과 : {100/x:.2f}')

except ValueError:
    print('유효한 정수가 아닙니다.')
except ZeroDivisionError: # 예외가 발생할 경우 except가 실행. 0 입력 시 '입력한 정수는 0 입니다.'는 출력됨.
    print('유효한 정수가 아닙니다.')
    
print('Done') # except 이후의 코드도 모두 실행됨.
````

````python
# 100을 나눌 유효한 정수를 입력할 때까지 정수를 입력받아 유효한 정수를 입력하면 입력한 정수와 100으로 나눈 수를 출력하기
while True:
    try:
        x = int(input('정수를 입력하세요 : '))
        print(f'입력한 정수는 {x} 입니다.')
        print(f'100을 {x}로 나눈 결과 : {100/x:.10f}')
        break

    except:
        print('유효한 정수가 아닙니다.')
````
