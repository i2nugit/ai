# 1. 조건문(if)

```
if 조건 1 :
    조건 1이 참일 경우 수행할 명령어들(if문 구절 단독 사용 가능, 들여쓰기 필수)
elif 조건 2 :
    조건 1이 거짓이고, 조건 2가 참일 경우 수행할 명령어들(필요한 만큼 추가 사용 가능)
else :
    모든 조건이 거짓일 경우 수행할 명령어들(단독으로 사용 불가; if 구절 필수)
```

````python
score = int(input('Enter your score > '))
if score>=60 :
    print("pass 입니다.")
else :
    print("fail 입니다.")
````

## 2.1 for-in문

```
for 임시 변수 in 나열 가능한 자료 :
```

````python
for i in range(1, 10, 2):
    print(i, end="\t")
else:
    print("\nDone")
````

````python
for i in range(1, 10):
    if i==5:
        break # 현재 수행 중인 반복문을 종료하는(빠져나오는) 명령어
    print(i, end="\t")
else:
    print('\nDone')
print('END')
````

## 2.2 while문
```
while 조건 :
    조건이 참일 경우 반복할 문장
else:
    조건이 거짓일 경우 한번 실행할 문장(else 구문 생략 가능)
```

````python
# 1~100 중 짝수들의 합을 출력하시오

tot = 0
i = 1

while i<=100:
    if i%2==0:
        tot += i
    i += 1 # if 구절이 종료될 때마다 i 값이 1 증가
    
print(f"{tot}")
````

````python
# 사용자에게 두 정수를 입력받아 작은수부터 큰수까지 홀수들의 합을 누적하여 출력하시오

fst_num = int(input('첫번째 정수 입력 : '))
sec_num = int(input('두번째 정수 입력 : '))
fst_number = fst_num
sec_number = sec_num

bigsmall = fst_num - sec_num
tot = 0

if bigsmall>0:
    while sec_num<=fst_num:
        if sec_num%2==1:
            tot += sec_num
        sec_num += 1
else:
    while fst_num<=sec_num:
        if fst_num%2==1:
            tot += fst_num
        fst_num += 1
        
        
print(f"{fst_number}, {sec_number} 사이 홀수들의 합은 {tot} 입니다.")
````

````python
first = int(input('첫번째 정수 입력 : '))
second = int(input('두번째 정수 입력 : '))

if first>second:
    first, second = second, first # 변수 바꾸기
    
tot = 0

for no in range(first, second):
    if no%2:
        tot += no
        
print(f"{first}, {second} 사이 홀수들의 합은 {tot} 입니다.")
````

````python
# 난수
import random
print(random.random()) #0부터 1 미만의 난수
print(random.randint(1,45)) #1부터 45 이하의 정수 난수
print(random.choice(range(1,46))) #1부터 46 미만의 정수 난수
print(random.choices(["A", "B"], weights=[.75, .25])) #A,B 중 하나, 단 각각의 확률은 75:25로 설정
print(random.choices(["A", "B", "C"], k=2)) #k개 복원 추출
print(random.choices(range(1,46), k=6)) #1부터 46 미만의 정수 중 중복 가능한 6개의 샘플을 수집
print(random.sample(range(1,46), 6)) #1부터 46 미만의 정수 중 중복되지 않은 6개의 샘플을 수집
````

````python
# 1~45 까지의 로또 번호 6 개를 추출하여 정렬된 로또번호를 출력하는 프로그램을 구현하시오

# 1 난수 생성 -> 리스트에 있는지 확인 -> 없을 경우 리스트의 아이템 개수가 6개 이하인지 확인 -> 리스트에 추가

lst = []
cnt = 0 #난수 생성 횟수
while len(lst)<6 :
    cnt += 1
    numb = random.randint(1,45)
    if numb not in lst: #numb가 lst에 없다면 아래 함수 수행
        lst.append(numb)
        
print(sorted(lst), cnt)
````

## 2.3 break, continue

- break : 반복문 진행 중, break 시 break를 포함하는 반복문을 종료
- continue : 반복문 내에서 continue 시 이후의 문장을 건너뜀

````python
num = 0
while num <=10:
    num += 1
    if num == 5:
        #break # 현재의 반복문 탈출
        continue # 현재 반복문의 조건으로 (출력 시 5가 없음)
    print(num, end="\t")
````

````python
# 10. 1~100 사이의 정수를 맞추는 프로그램을 구현하시오
'''
프로그램 안에서 random 수가 만들어지고, 그 random 수 맞추기

1단계 : 입력된 숫자가 틀리면 UP/DOWN 메시지를 출력하고 숫자를 맞추면 축하메세지와 함꼐 종료합니다 맞출 때까지 반복
2단계 : 수를 맞추는데 실패하면 min/max를 조정, 맞추면 축하메세지와 함께 종료
3단계 : 2단계의 기회를 5회로 제한. 5회의 기회가 지나면 실패메세지와 함께 종료
'''
````

````python
comnum = random.randint(1,100)
print(f"컴퓨터가 생성한 임의의 수 {comnum}")

minnum = 1
maxnum = 100
i = 0

while i <= 5:
    i += 1
    j = 5 - i
    inputnum = int(input(f"{minnum}~{maxnum} 사이의 수를 맞춰보세요 : "))
    if inputnum == comnum:
        print("축하드립니다.")
        break
    elif i == 5:
        print("기회를 모두 소진하였습니다.")
        break
    elif (inputnum > maxnum) | (inputnum < minnum):
        print(f"수의 범위를 벗어났습니다. \n남은 기회는 {j}번 입니다.")
    elif inputnum > comnum:
        maxnum = inputnum
        print(f"DOWN ↓ \n남은 기회는 {j}번 입니다.")
    elif inputnum < comnum:
        minnum = inputnum
        print(f"UP ↑ \n남은 기회는 {j}번 입니다.")
````

# 4. 중첩 루프 탈출

````python
for a in range(0, 3):
    for b in range(1, 3):
        if a==b:
            print(f"{a}, {b} : break")
            break
        print(f"{a}, {b}")
        
print('done')
````

````python
# 중첩 루프 탈출 방법 1 : flag 변수 활용

flag = False

for a in range(0, 3):
    for b in range(1, 3):
        if a==b:
            print(f"{a}, {b} : break")
            flag = True
            break
        print(f"{a}, {b}")
    if flag:
        break
    
    
    
print('done')
````
