## 1.3 지역변수(local var)와 전역변수(global var)
- 변수의 값을 조회할 때, 로컬 심볼 테이블 - 전역 심볼 테이블 - 내장 이름 테이블 순서로 조회한다.

````python
global_val = 100 # 전역변수

def func1():
    print('함수 안 :', global_val)
func1()
print('함수 밖 :', global_val)
````

````python
def func2():
    local_var = 200 #지역변수
    print('함수 안 :', local_var)
func2()
print('함수 밖 :', local_var)
````

````python
# 변수의 값을 조회하는 순서 : 로컬 심볼 테이블 - 전역 심볼 테이블 - 내장 이름 테이블

g_var = 100 #전역변수

def func3():
    g_var = 200 #지역변수
    print("함수 안 :", g_var) 

func3()
print("함수 밖 :", g_var) 
````

````python
# 렉시컬 특성 : 변수의 선언 위치에 따라 변수 범위가 정해짐
# -> 전역변수로 값이 잡혀있더라도, 함수 내에서 지역변수로 새로 지정되면 전역변수와 관련 없이 따로 할당이 필요함

g_var = 100 #전역변수

def func3():
    print("함수 안 :", g_var) # g_var가 지역변수로 지정되었으나 변수의 값이 할당되지 않음(초기화 안됨)[처리 순서 2]
    g_var = 200 #지역변수로 지정 [처리 순서 1]
func3()
print("함수 밖 :", g_var) 
````

````python
g_var = 100 #전역변수

def func3():
    global g_var # func3 함수 내의 g_var는 전역변수를 사용함을 선언
    print("함수 안 :", g_var) #전역변수
    g_var = 200 #전역변수
func3()
print("함수 밖 :", g_var) 
````

## 1.4 값에 의한 호출
- 함수의 인수로 전달되는 변수가 스칼라 변수일 때, 해당 변수의 id는 변하지 않음

````python
foo = 'a' # 전역변수
id(foo)
````

````python
def func1(f): # 매개변수 f는 지역변수
    print("before :", f, '의 주소는',id(f))
    f = f*3
    print("after :", f, '의 주소는',id(f))
    
func1(foo) # foo의 주소는 그대로, 함수의 결과값의 주소만 변경됨.
print('foo : a 의 주소는 :', id(foo))
````

## 1.5 참조에 의한 호출
- 함수의 인수로 전달되는 변수가 리스트, 튜플, 셋, 딕셔너리, 객체일 때, 해당 변수의 id는 변하지 않음

````python
lst = [1, 2, 3, 4, 5] # 전역변수
id(lst)
````

````python
def func1(f): # 매개변수 f는 지역변수
    print("before :", f, '의 주소는',id(f))
    f.append(6)
    print("after :", f, '의 주소는',id(f))
    
func1(lst) # foo의 주소는 그대로, 함수의 결과값의 주소만 변경됨.
print('lst :', lst, '의 주소는', id(lst))
````

# 2. return
- 함수의 실행 결과를 반환함
- 반환 값을 함수의 결과를 호출한 위치에 전달하기 위해 사용

````python
def my_add_print(a, b): #return이 없는 함수
    "매개변수로 들어온 a, b 값의 합을 출력합니다"
    total = a + b
    print(f"합은 {total}입니다.")
    
def my_add_return(a, b): #return이 있는 함수
    "매개변수로 들어온 a, b 값의 합을 반환합니다"
    total = a + b
    return f"합은 {total}입니다."
````

````python
result = my_add_print(1, 2)
print(result)
print(type(result))
````

````python
result = my_add_return(1, 2)
print(result)
print(type(result))
````

## 여러 값을 반환

````python
def swap(a, b): #return이 있는 함수
    "매개변수로 들어온 a, b 순서를 바꾸어 반환합니다"
    return b, a #여러 값을 return할 경우, tuple로 전달됨
````

````python
result = swap(1, 2)
````

````python
print(result)
print(type(result))
````

````python
x, y = 10, 20
print(f"함수 실행 전 : x = {x}, y = {y}")
x, y = swap(x, y) #함수가 반환하는 값의 개수만큼 변수를 선언하여 각각의 변수가 반환 값을 받도록 할 수 있음
print(f"함수 실행 후 : x = {x}, y = {y}")
````

# 3. 함수의 매개변수 ⭐
## 3.1 기본값을 갖는 매개변수

````python
def make_url (ip, port=80): #ip:순서인수, port:기본값을 갖는 매개변수; 입력이 없을 때 80의 기본값을 가진다
    return f"http://{ip}:{port}"
````

````python
print(make_url('www.naver.com',90)) #입력한 순서에 따라 함수의 변수에 입력됨
print(make_url(90, 'www.naver.com'))
print(make_url(port=9090, ip='www.naver.com')) #키워드 인수
````

````python
def make_url (port=80, ip): #순서인수가 있을 때, 기본값을 갖는 매개변수의 뒷순서로 갈 경우 에러 발생
    return f"http://{ip}:{port}"
````

````python
print(make_url('www.naver.com',90)) #입력한 순서에 따라 함수의 변수에 입력됨
print(make_url(90, 'www.naver.com'))
print(make_url(port=9090, ip='www.naver.com')) #키워드 인수
````

## 3.3 순서인수, 키워드인수
- 순서인수 : 함수에 정의(입력)된 순서대로 전달하는 방식
- 키워드인수 : 인수의 이름을 명시하는 방식

````python
def func(a, L=None):
    if L is None:
        L = []
    L.append(a)
    return L
````

````python
lst = []
print(func(10, lst)) #순서인수
print(func(L=[], a=20)) #키워드인수
print(func(20)) #순서인수 a에 20을 전달, L의 기본값이 None이므로 생략 가능
````
