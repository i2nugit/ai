## 1.3 지역변수(local var)와 전역변수(global var)
- 변수의 값을 조회할 때, 로컬 심볼 테이블 - 전역 심볼 테이블 - 내장 이름 테이블 순서로 조회한다.

````python
global_val = 100 # 전역변수

def func1():
    print('함수 안 :', global_val)
func1()
print('함수 밖 :', global_val)
````

````python
def func2():
    local_var = 200 #지역변수
    print('함수 안 :', local_var)
func2()
print('함수 밖 :', local_var)
````

````python
# 변수의 값을 조회하는 순서 : 로컬 심볼 테이블 - 전역 심볼 테이블 - 내장 이름 테이블

g_var = 100 #전역변수

def func3():
    g_var = 200 #지역변수
    print("함수 안 :", g_var) 

func3()
print("함수 밖 :", g_var) 
````

````python
# 렉시컬 특성 : 변수의 선언 위치에 따라 변수 범위가 정해짐
# -> 전역변수로 값이 잡혀있더라도, 함수 내에서 지역변수로 새로 지정되면 전역변수와 관련 없이 따로 할당이 필요함

g_var = 100 #전역변수

def func3():
    print("함수 안 :", g_var) # g_var가 지역변수로 지정되었으나 변수의 값이 할당되지 않음(초기화 안됨)[처리 순서 2]
    g_var = 200 #지역변수로 지정 [처리 순서 1]
func3()
print("함수 밖 :", g_var) 
````

````python
g_var = 100 #전역변수

def func3():
    global g_var # func3 함수 내의 g_var는 전역변수를 사용함을 선언
    print("함수 안 :", g_var) #전역변수
    g_var = 200 #전역변수
func3()
print("함수 밖 :", g_var) 
````

- 반환 값을 함수의 결과를 호출한 위치에 전달하기 위해 사용

````python
def my_add_print(a, b): #return이 없는 함수
    "매개변수로 들어온 a, b 값의 합을 출력합니다"
    total = a + b
    print(f"합은 {total}입니다.")
    
def my_add_return(a, b): #return이 있는 함수
    "매개변수로 들어온 a, b 값의 합을 반환합니다"
    total = a + b
    return f"합은 {total}입니다."
````

## 3.3 순서인수, 키워드인수
- 순서인수 : 함수에 정의(입력)된 순서대로 전달하는 방식
- 키워드인수 : 인수의 이름을 명시하는 방식

````python
def func(a, L=None):
    if L is None:
        L = []
    L.append(a)
    return L
````

````python
#함수 정의

def add(a, b):
    return a + b

def add(a, b, c):
    return a + b + c
````

````python
# *args의 * : 인수들이 args로 튜플 패킹되어 전달됨을 표기

def add(*args):
    "전달받은 인수들의 합을 반환. 매개변수가 없을 경우 0을 반환."
    # print(args) = (10,20)/(10, 2, 3)/ ...
    tot = 0 # 누적변수
    for num in args:
        tot += num
    return tot
````

````python
print(add())
print(add(10,20))
print(add(10,2,3))
````

## 3.7 인수 언패킹
### 튜플 인수 언패킹
- 패킹 : 함수 정의 시 매개변수 앞 * (튜플 매개변수)
- 언패킹 : 함수 호출 시 매개변수 앞 * (리스트나 튜플의 각자 요소별 매개변수로)

````python
def add(*args):
    tot = 0
    for num in args:
        tot += num
    return tot
````

## 4.2 매개변수에 람다식 사용
- 람다식을 사용하는 곳 : 함수의 매개변수, 람다식을 return
### map()

````python
def mapping(func, L=[]):
    result = []
    for item in L:
        result.append(func(item))
    return result
````

````python
lst = [1, 2, 3, 4, 5, 6, 7, 8]
mapping(lambda x:x*2, lst)
````

### filter()

````python
def filter_tmplt(func, L):
    result = []
    for item in L:
        if func(item):
            result.append(item)
    return result
````

````python
#4
def mysum(fRom, end):
    sum=0
    if fRom < end:
        for i in range(fRom, end+1):
            sum += i
    else:
        for i in range(end, fRom+1):
            sum += i
    return sum

mysum(1, 100)
````
