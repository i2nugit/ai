```python
def safe_index(lst, item):
    '''
    첫번째 매개변수 lst에서 두번째 매개변수 item이 있는 index를 반환하는 함수
    item이 없으면 -1을 반환
    '''
    if item in lst:
        return lst.index(item)
    return -1
```

# 5. enumerate
- 반복자 또는 순서 객체로 반복문을 처리할 때 사용하는 함수
- 리스트, 튜플, 셋을 for문에 적용 시 사용

```python
list(enumerate(names))
```

```python
for no, name in enumerate(names):
    print(f'{no} : {name}')
```

```python
names = ['길동', '길자','순자']
tels = ['010-0000-0000', '010-1111-1111', '010-2222-2222']
adds = ['강남', '인천', '판교']

for idx in range(len(names)): # range(len(names)) = [0, 1, 2]
    print ("{}. 이름 : {}, 연락처 : {}, 지역 : {}". format(idx+1, names[idx], tels[idx], adds[idx]))
```

```python
list(zip(names, tels, adds))
```

```python
for name, tel, add in zip(names, tels, adds):
    print(f"이름 : {name}, 연락처 : {tel}, 지역 : {add}")
```

```python
list(enumerate(zip(names, tels, adds)))
```

```python
for idx, (name, tel, add) in enumerate(zip(names, tels, adds)):
    print(f"{idx+1}. 이름 : {name}, 연락처 : {tel}, 지역 : {add}")
```

```python
# names, tels, adds를 딕셔너리 리스트로 만들기
members = []
for name, tel, add in zip(names, tels, adds):
    member = {'name':name, 'tel':tel, 'address':add}
    members.append(member)
members
```

```python
members = []
for idx, (name, tel, add) in enumerate(zip(names, tels, adds)):
    member = {'순서':idx+1,'name':name, 'tel':tel, 'address':add}
    members.append(member)
members
```
