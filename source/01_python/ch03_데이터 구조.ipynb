```python
# list의 item 출력하기
fruits = ['mango', 'apple', 'orange', 'grape']
for idx in range(len(fruits)): # [0, 1, 2, ... , n] ; n = fruits의 len() 결과값(list의 item 수)
    print("{}번째 과일은 {}".format(idx, fruits[idx]))
```

## 1.4 인덱싱과 item 인덱스 찾기
- 변수[idx] : 인덱스를 이용한 접근 (항목을 찾지 못할 경우 에러 발생)
- count(찾을 데이터) : 해당 데이터의 개수 반환
- index(찾을 데이터) : 0번째부터 해당 데이터의 index 위치 값을 반환
- index(찾을 데이트, start) : start번째부터 해당 데이터의 index 위치 값을 반환

```python
numbers = [1, 2, 3, [1, 2, 3], 1, 2, 3]
```

```python
numbers.count(9)
```

```python
print(2 in numbers) # 2가 numbers의 item인지 여부 확인
print(9 in numbers)
```

```python
numbers.index(2, 2)
```

```python
# index()는 에러 발생 가능성이 높아 count와 주로 함께 쓰임

if numbers.count(9):
    print(numbers.index(9))
else:
    print('Not found')
```

```python
if 9 in numbers:
    print(numbers.index(9))
else:
    print('Not found')
```

```python
def safe_index(lst, item):
    '''
    첫번째 매개변수 lst에서 두번째 매개변수 item이 있는 index를 반환하는 함수
    item이 없으면 -1을 반환
    '''
    if item in lst:
        return lst.index(item)
    return -1
```

## 1.5 특정 item 추출
1. 리스트 슬라이싱 : list에서 특정 index의 데이터를 추출(인덱스 범위를 벗어나도 에러 발생 x)
 - [from:stop:step] : from 부터 stop 전까지 step 씩 전진하며 index 값 추출
     * from 생략 시 : 처음부터, stop 생략 시 : 끝까지, step 생략 시 1, 모두 생략 가능
     * 인덱스 음수 가능 <br><br>
2. <font color='red'>리스트 컴프리헨션</font> : list에서 원하는 조건의 데이터로 새로운 리스트 생성

```python
numbers = [1, 3, 5, 9, 10, 12, 13, 18]
# 거꾸로 출력, 짝수번째 출력, 모두 출력 (슬라이싱 이용)

print(numbers[::-1])
print(numbers[2::2])
print(numbers[::])
```

```python
# 홀수만 추출
# lst = []
# for item in numbers:
#     if item%2:
#         lst.append(item)


lst = [item for item in numbers if item%2]
lst
```

```python
# 짝수만 추출

lst = [item for item in numbers if item%2==0]
lst
```

```python
# 10 이상만 추출

over10 = [item for item in numbers if item>=10]
over10
```

```python
# fruits 리스트에서 과일 글자수를 추출
fruits = ['orange', 'apple', 'mango','kiwi']

lst = [len(a) for a in fruits]
lst
```

```python
# 과일의 글자가 5글자 이상인 과일만 추출

over5 = [n for n in fruits if len(n)>=5]
over5
```

```python
# lst에서 3이상인 요소들의 제곱을 추출
lst = [1, 2, 3, 4, 5]

over3 = [m**2 for m in lst if m>=3]
over3
```

```python
# lst의 두배값을 추출

dob = [b*2 for b in lst]
dob
```

## 1.7 item 삭제

```python
# pop() : 맨 뒤의 항목 반환 및 삭제
# pop(idx) : idx 위치의 항목 반환 및 삭제
# del(list변수[idx]) : idx 위치의 항목 삭제
# remove(data) : list 내 해당 data 항목 삭제 (data가 item이 아닐 경우 에러 발생)
```

```python
numbers = [1, 2, 3, 4, 5]
print('삭제한 데이터 :', numbers.pop())
print("삭제 후 numbers :", numbers)
print('삭제한 데이터 :', numbers.pop(0))
print("삭제 후 numbers :", numbers)
del(numbers[1])
print("삭제 후 numbers :", numbers)
numbers.remove(4)
print("삭제 후 numbers :", numbers)
```

## 1.9 list 복제

```python
# 스칼라 변수(변수 하나당 하나의 데이터)의 복제
i = 10
copied_i = i
print(f"원본 i = {i}, 복사본 i = {copied_i} / 원본 주소 : {id(i)}, 복사본 주소 : {id(copied_i)}")
copied_i +=5
print(f"원본 i = {i}, 복사본 i = {copied_i} / 원본 주소 : {id(i)}, 복사본 주소 : {id(copied_i)}")
```

```python
# 데이터 구조의 복제
numbers = [5, 1, 9]   # 원본
new_numbers = numbers # 복제본
print(f"원본 = {numbers}, 복사본 = {new_numbers} / 원본 주소 : {id(numbers)}, 복사본 주소 : {id(new_numbers)}")
```

```python
numbers = [5, 1, 9] 
new_numbers = numbers # 복제본 x, 할당 o
new_numbers.sort()
print(f"원본 = {numbers}, 복사본 = {new_numbers} / 원본 주소 : {id(numbers)}, 복사본 주소 : {id(new_numbers)}")
```

```python
# 복제 (= 깊은 복사)
numbers = [5, 1, 9]          # 원본
new_numbers = numbers.copy() # 복제본
new_numbers.sort()
print(f"원본 = {numbers}, 복사본 = {new_numbers} / 원본 주소 : {id(numbers)}, 복사본 주소 : {id(new_numbers)}")
```
