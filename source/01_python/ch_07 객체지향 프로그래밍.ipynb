# 2. 생성자와 소멸자   
```
- 생성자 '__init__(self, 매개변수, ...)' : 객체가 생성될 때 자동 실행
                                         생성 시 필요한 코드(인스턴스 변수 초기화) 포함
                                         생성자가 없을 경우 자동 생성

- 소멸자 '__del__(self)' : 객체가 소멸될 때 자동 실행 (인스턴스 객체의 래퍼런스 카운트가 0이 될 때)
                          소멸 시 필요한 코드 포함

    - self : 이 함수를 호출한 객체(인스턴스) 자기 자신, 전 세계적인 약속(PEP 8 규칙)이므로 self로 작성 
```

````python
class Person:
    def __init__(self, name, gender):
        self.name = name
        self.gender = gender
        print('Person 생성자 호출(=인스턴스 메모리 생성, 객체 생성, 클래스형 변수 생성, 타입 생성)')
    def __del__(self):
        print('Person 객체 소멸')
    def print_info(self):
        print(f'{self.name}님은 {self.gender}입니다.')
    def __str__(self):
        return f'{self.name}님은 {self.gender}입니다'
````

````python
p1 = Person('심사임당', '여자')
p1.print_info()
print(p1.__str__())
````

````python
type(p1), isinstance(p1, Person)
````

````python
del p1 # 래퍼런스 카운트 : 0
````

````python
p2 = Person('율곡이이', '남자') # 래퍼런스 카운트 : 0 → 1
p3 = p2 # 래퍼런스 카운트 : 1 → 2
p2.print_info()
````

````python
del p2 # 래퍼런스 카운트 2 → 1
````

````python
del p3 # 래퍼런스 카운트 1 → 0 ; 소멸자 실행
````

````python
class Person:
    # 생성자를 이용한 인스턴스 변수 초기화
#     def __init__(self): 
#         self.name = '익명'
#         self.gender = '남성' ; 아래의 init에 의해 비활성화됨
    def __init__(self, name):
        self.name = name
        self.gender = '남성'
    def __str__(self):
        return f'{self.name}님은 {self.gender} 입니다.'
````

````python
p1 = Person()
p1 = Person('홍길동')
print(p1)
````

````python
class Person:
    # 생성자 함수는 중복 정의가 불가하므로, 기본값을 갖는 매개변수를 이용해 생성자를 정의한다
    def __init__(self, name = '익명', gender = '남성'):
        self.name = name
        self.gender = gender
    def __str__(self):
        return f'{self.name} 님은 {self.gender} 입니다.'
````

````python
p1 = Person()
print(p1)
p1 = Person('홍길동')
print(p1)
p1 = Person('박금자', '여성')
print(p1)
````

# 3. 변수와 메소드(함수)
- 인스턴스 변수(멤버 변수) / 클래스 변수 / 스태틱 변수(정적 변수)
- 인스턴스 메소드(self) / 클래스 메소드(cls) / 스태틱 메소드()

````python
# 클래스 변수 : 클래스 내부 선언 → 생성된 모든 인스턴스에 동일한 값을 가짐 
#      [클래스].[클래스 변수] 수정 시모든 인스턴스에 반영될 수 있음
#     [객체변수].[클래스변수] 수정 시 해당 객체에만 반영됨

# static 변수 : 클래스 내부 선언. static 변수 앞에 언더바 2개(__) 붙임
````

````python
class Person:
    academy = 'MBC' # 클래스 변수
    def __init__(self, name = '홍길동', gender = '남성'):
        self.name = name
        self.gender = gender
        print('Person 객체 생성합니다')
    def __str__(self):
        return "%s의 %s은/는 %s 입니다."%(self.academy, self.name, self.gender)
````

````python
print(Person)
print(Person.academy)
````

````python
p1 = Person()
p2 = Person()
print(p1)
print(p2)
````

````python
Person.academy = 'KBS'
p2.academy = 'SBS'
p2.name = '마동석'
print(p1)
print(p2)
````
