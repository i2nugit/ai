# 2. 생성자와 소멸자   
```
- 생성자 '__init__(self, 매개변수, ...)' : 객체가 생성될 때 자동 실행
                                         생성 시 필요한 코드(인스턴스 변수 초기화) 포함
                                         생성자가 없을 경우 자동 생성

- 소멸자 '__del__(self)' : 객체가 소멸될 때 자동 실행 (인스턴스 객체의 래퍼런스 카운트가 0이 될 때)
                          소멸 시 필요한 코드 포함

    - self : 이 함수를 호출한 객체(인스턴스) 자기 자신, 전 세계적인 약속(PEP 8 규칙)이므로 self로 작성 
```

````python
class Person:
    def __init__(self, name, gender):
        self.name = name
        self.gender = gender
        print('Person 생성자 호출(=인스턴스 메모리 생성, 객체 생성, 클래스형 변수 생성, 타입 생성)')
    def __del__(self):
        print('Person 객체 소멸')
    def print_info(self):
        print(f'{self.name}님은 {self.gender}입니다.')
    def __str__(self):
        return f'{self.name}님은 {self.gender}입니다'
````

````python
p1 = Person('심사임당', '여자')
p1.print_info()
print(p1.__str__())
````

````python
del p1 # 래퍼런스 카운트 : 0
````

# 3. 변수와 메소드(함수)
- 인스턴스 변수(멤버 변수) / 클래스 변수 / 스태틱 변수(정적 변수)

              [클래스].[클래스 변수] 수정 시모든 인스턴스에 반영될 수 있음

※ class 내의 메소드(함수) 종류
- 인스턴스 메소드 (일반 함수) : 첫번째 매개변수에 self를 꼭 명시. 객체를 통해서만 호출할 메소드.
- 클래스 메소드 : 메소드 위에 @classmethod가 선언된 메소드. 첫 번째 매개변수로 클래스 자신을 가리키는 cls를 받음.
- static 메소드 : 메소드 위에 @staticmethod가 선언된 메소드. 디폴트 매개변수가 없음.

````python
class Person:
    academy = 'MBC' # 클래스 변수
    def __init__(self, name='홍길동',gender='남자'): # 생성자
        self.name = name
        self.gender = gender
        
        
    # 1. 인스턴스 메소드: 객체의 상태를 출력하거나 변경할 때   
    
    def print_info(self): #변수.print_info()로 호출 가능
        print(f"인스턴스 메소드 - {self.academy}의 {self.name}은(는) {self.gender}입니다.")
    def change_info(self, name, gender='남자'): # 변수.change_info('신길숙','여자')로 호출 가능
        self.name = name
        self.gender = gender
        print(f"인스턴스 메소드를 통해 {self.name}(으)로 이름 변경, {self.gender}로 성별 변경 완료")
 

    # 2. 클래스 메소드: 클래스 변수(공유 데이터)를 다룰 때
    
    @classmethod # 바로 아래의 영역만 해당됨
    def change_class_method(cls, academy): # cls == Person ; 변수.change_class_method("KBS")로 호출 가능
        cls.academy = academy # 클래스 영역의 클래스 변수 수정
        print(f"클래스 메소드 - {cls.academy}로 클래스 변수 변경")

        
    # 3. 스태틱 메소드: 클래스/인스턴스 데이터와 무관한 독립적 로직일 때
    
    @staticmethod
    def print_static_method():
        print(f'static 메소드 : {Person.academy}로 유지')
        
    @staticmethod
    def is_valid_gender(gender_input): # 입력한 성별이 정확한 정보인지 확인
        # self나 cls를 받지 않고, 들어온 매개변수만 가지고 판단하는 '심부름꾼' 역할
        valid_list = ['남자', '여자', '남', '여', '그 외']
        return gender_input in valid_list
````

````python
p1 = Person()
p1.print_info()
````

````python
Person.change_class_method('KBS')
Person.print_static_method()
p1.print_info()
````

````python
p1.academy = '신림 MBC'
p1.change_info('김길숙','여자')
p1.print_info()
````

````python
p2 = Person()
p2.print_info()
````

- override(함수의 재정의) - 가능 ; 부모 클래스의 함수를 자식 클래스에서 재정의. 함수 이름, 매개변수 개수 동일.
- overload(함수의 중복 정의) - 불가능 ; 동일 이름의 함수에 매개변수를 달리하여 정의.

````python
class Person:
    def __init__(self, name, gender):
        print("Person 객체 생성")
        self.name = name
        self.gender = gender
    def __del__(self):
        print("Person 객체 소멸")
    def print_info(self):
        print(self) # = __str__()이 호출되어 f'{self.name} 님은 {self.gender} 입니다.'가 출력됨
    def __str__(self):
        return f'{self.name} 님은 {self.gender} 입니다.'
````

````python
class Student(Person):
    def __init__(self, name, gender, major): # __init__ 재정의
        print("Person-Student 객체 생성")
#         self.name = name
#         self.gender = gender
        Person.__init__(self, name, gender) # Person의 init을 가져옴
        self.major = major
    def __del__(self): # __del__ 재정의
        pass
    def __str__(self):
        return f'{self.name} 님은 {self.gender}이고, {self.major} 전공 입니다.'
````

## super()
- super() : '부모 클래스의'를 의미

````python
class Person:
    def __init__(self, name, gender):
        print("Person 객체 생성")
        self.name = name
        self.gender = gender
    def __del__(self):
        print("Person 객체 소멸")
    def print_info(self):
        print(self) # = __str__()이 호출되어 f'{self.name} 님은 {self.gender} 입니다.'가 출력됨
    def __str__(self):
        return f'{self.name} 님은 {self.gender} 입니다.'
````

````python
class Student(Person):
    def __init__(self, name, gender, major): # __init__ 재정의
        print("Person-Student 객체 생성")
        # Person.__init__(self, name, gender) : Person의 init을 가져옴
        super().__init__(name, gender) # 부모 클래스(Person)의 __init__을 name과 gender를 통해 결과값 반환
        self.major = major
    def __del__(self): # __del__ 재정의
        pass
    def print_info(self):
        super().print_info()
        print('전공은 {} 입니다.'.format(self.major))
    def __str__(self):
        return super().__str__() + f' 그리고 {self.major} 전공 입니다.'
````

````python
s2 = Student("홍탁순", "남자", "체육교육학")
````

````python
print(s2)
s2.print_info()
````

## static 변수 (정적 변수)
- static 변수 : 클래스로 만든 모든 객체(인스턴스)들이 다 함께 공유하는 하나의 변수
* 네임 맹글링 : 파이썬에서 클래스 변수나 인스턴스 변수 이름 앞에 `__`(언더스코어 2개)를 붙이면, 파이썬이 내부적으로 그 이름을 `_클래스명__변수명` 형태로 변경함
* 클래스 변수와 인스턴스 변수는 외부에서 접근 가능. 반면 `__`로 시작하는 변수의 경우, 외부에서 `클래스명._클래스명__변수명`으로 접근이 필요해 실수로 인한 변경을 방지
    - 외부에서 실수로 변수를 건드리거나, 상속받은 자식 클래스에서 변수 이름이 겹쳐서 덮어쓰기(오버라이딩)되는 사고를 방지

````python
class Person:
    def __init__(self, name, gender):
        print("Person 객체 생성")
        self.name = name
        self.gender = gender
    def __del__(self):
        print("Person 객체 소멸")
    def print_info(self):
        print(self)
    def __str__(self):
        return f'{self.name} 님은 {self.gender} 입니다.'
    
class Student(Person):
    __count = 0 # static 변수 (정적 변수,공유 변수) ; Student형 변수 생성 개수를 저장
    def __init__(self, name, gender, major):
        Student._Student__count += 1 # 객체가 생성될 때마다 __count 1 증가
        super().__init__(name, gender)
        self.major = major
    def __del__(self):
         Student._Student__count -= 1 # 객체가 소멸될 때마다 __count 1 감소
    def print_info(self):
        super().print_info()
        print('전공은 {} 입니다.'.format(self.major))
        print('{} 명 등록됨.'.format(Student._Student__count))
    def __str__(self):
        return super().__str__() + f' 그리고 {self.major} 전공 입니다. 총 {Student._Student__count} 명 등록되었습니다.'
    
    @classmethod
    def get_count(cls):
        return cls.__count
````

````python
s1 = Student("홍길동", "남자", "AI")
print(s1)
print(Student.get_count())
print(s1.get_count())
````

````python
Student.__count = 0 # 외부에서 __count로의 접근 불가
Student.get_count()
````

````python
s2 = Student("고말숙", "여자", "컴퓨터공학")
print(s2)
print(Student.get_count())
````

````python
del s2
print(Student.get_count())
````

````python
print(Student.__dict__) # __count가 _Student__count로 바뀌어있음
````

# 연습문제
## 실습형

````python
class Shape:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
        
    def move(self, a, b):
        self.x = self.x + a
        self.y = self.y + b
        
    def __str__(self):
        return f'{self.x}, {self.y}'
    
    @staticmethod
    def is_valid_number(number_input):
        if isinstance(number_input, int):
            print('사용 가능한 정수입니다.')
        else:
            print('정수가 아닙니다.')
````

````python
s1 = Shape(5, 5)
Shape.is_valid_number(22.1)
s1.move(10, 5)
print(s1)
````

````python
class Triangle(Shape):    
    __count = 0    
    def __init__(self, width, height, x, y):
        print('생성')
        Triangle._Triangle__count += 1
        super().__init__(x, y)
        self.width = width
        self.height = height
        
    def __del__(self):
        Triangle._Triangle__count -= 1
        
    def calc_area(self):
        print(f'{self.width * self.height / 2}')
        
    def __str__(self):
        return f'좌표 : {self.x}, {self.y}, 밑면 : {self.width}, 높이 : {self.height}, 넓이 : {self.width * self.height / 2}'
    
    @classmethod
    def get_count(cls):
        return cls.__count
````

````python
t1 = Triangle(4, 3, 0, 10)
print(t1)
print('Triangle 객체 개수 :', Triangle.get_count())
t1.move(50,50)
print(t1)
````

````python
t2 = Triangle(3, 4, 5, 5)
print(t2)
print('Triangle 객체 개수 :', Triangle.get_count())
print(t2)
````

````python
del t1, t2
print('Triangle 객체 개수 :', Triangle.get_count())
````
